```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
def contrast(img, a, s, t):        
    img = a * (img + s) - t
    
    img = np.clip(img, 0 , 255)
    
    return img.astype('uint8')
```

```python
images = ['low_contrast.jpg', 'high_contrast.jpg', 'low_detalization.jpg', 'high_detalization.jpg']
for orig_img in images:
    orig_img = Image.open(orig_img).convert('RGB')
    
    img = np.asarray(orig_img).astype('int64')

    low_contrast_image = contrast(img, 0.6, 100, 2)

    fig = plt.figure()
    fig.set_figheight(20)
    fig.set_figwidth(20)

    fig.add_subplot(1, 2, 1)
    plt.imshow(orig_img)

    fig.add_subplot(1, 2, 2)
    plt.imshow(low_contrast_image)
```
