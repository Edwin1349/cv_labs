```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy import spatial
```

```python
def BFMatcher(kp1, des1, kp2, des2):# create BFMatcher object 
    bf = cv.BFMatcher(cv. NORM_L2, crossCheck=True)
    
    matches = bf.match(des1,des2) 
    matches = sorted(matches, key = lambda x:x.distance) 
    
    img = cv.drawMatches(img1,kp1,img2,kp2,matches[:10],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    
    return img
```

```python
def CustomMatcher(kp1, des1, kp2, des2):
    matches = []
    dist = cdist(des1, des2, 'euclidean').flatten()
    query = [*range(des1.shape[0])] * des2.shape[0]
    train = [*range(des2.shape[0])] * des1.shape[0]
    print(query)
    print(train)
    
    for i in range(len(dist)):
        matches.append(cv.DMatch(_distance=dist[i], _imgIdx=0, _queryIdx=query[i], _trainIdx=train[i]))
            
    matches = sorted(matches, key = lambda x : x.distance)
    img = cv.drawMatches(img1, kp1, img2, kp2, matches[:10], None, flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    
    return img
```

```python
queries = ['./image1.jpg', './image2.jpg']
trains = ['./test1.jpg', './test2.jpg']
for i in range(len(queries)):
    img1 = cv.imread(queries[i],cv.IMREAD_GRAYSCALE)
    img2 = cv.imread(trains[i],cv.IMREAD_GRAYSCALE)

    sift = cv.SIFT_create() 
    kp1, des1 = sift.detectAndCompute(img1,None) 
    kp2, des2 = sift.detectAndCompute(img2,None)

    fig = plt.figure()
    fig.set_figheight(40)
    fig.set_figwidth(40)

    fig.add_subplot(1, 2, 1)
    plt.imshow(BFMatcher(kp1, des1, kp2, des2))

    fig.add_subplot(1, 2, 2)
    plt.imshow(CustomMatcher(kp1, des1, kp2, des2))
```
